```python
import pandas as pds

journal_dict = dict()

js_file = open("../servier/de/pub_med/data/output/graph_drugs.json", 'r')

df = pds.read_json(js_file.read())

for papers in df['map'].apply(lambda x : x['journal']).to_list():
    for p in papers:
        try:
            journal_dict[p['title']] += 1
        except:
            journal_dict[p['title']] = 1

df = pds.DataFrame(journal_dict.items(), columns=['journal', 'score'])

df.plot.bar(x='journal', y='score', rot=90)
```
